```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------
# 1. Create Dataset
# -----------------------------------
np.random.seed(42)
m = 50
X = 2 * np.random.rand(m, 1)
y = 4 + 3 * X + np.random.randn(m, 1)

# -----------------------------------
# 2. Model Function
# f(x) = wx + b
# -----------------------------------
def model(X, w, b):
    return w * X + b

# -----------------------------------
# 3. Cost Function (MSE)
# J(w,b) = (1/2m) * sum((prediction - y)^2)
# -----------------------------------
def compute_cost(X, y, w, b):
    m = len(y)
    predictions = model(X, w, b)
    cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
    return cost

# -----------------------------------
# 4. Gradient Descent
# -----------------------------------
def gradient_descent(X, y, w, b, alpha, iterations):
    m = len(y)
    cost_history = []

    for i in range(iterations):
        predictions = model(X, w, b)

        dw = (1 / m) * np.sum((predictions - y) * X)
        db = (1 / m) * np.sum(predictions - y)

        w = w - alpha * dw
        b = b - alpha * db

        cost = compute_cost(X, y, w, b)
        cost_history.append(cost)

    return w, b, cost_history

# -----------------------------------
# 5. Training
# -----------------------------------
w = 0
b = 0
alpha = 0.1
iterations = 100

w_final, b_final, cost_history = gradient_descent(X, y, w, b, alpha, iterations)

print("Final w:", w_final)
print("Final b:", b_final)
print("Final Cost:", compute_cost(X, y, w_final, b_final))

# -----------------------------------
# 6. Plot Cost
# -----------------------------------
plt.plot(cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Decreasing Over Time")
plt.show()

# -----------------------------------
# 7. Plot Regression Line
# -----------------------------------
plt.scatter(X, y)
plt.plot(X, model(X, w_final, b_final), color="red")
plt.title("Linear Regression Result")
plt.show()
```
