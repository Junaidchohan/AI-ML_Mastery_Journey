# Linear Regression Cost Function

## Goal
Build and understand Linear Regression using only basic Python and NumPy.

We will:

- Define a simple linear model
- Implement the cost function manually
- Visualize predictions
- Observe how cost changes with different parameters

No external ML libraries will be used.

## Problem Statement

We want to predict house prices based on house size.

### Training Data

| Size (1000 sqft) | Price (1000s dollars) |
|------------------|-----------------------|
| 1                | 300                   |
| 2                | 500                   |

### Model

We use a linear model:

f(x) = w*x + b

Where:
- w = slope
- b = intercept
- x = input (house size)
- f(x) = predicted price

```python
import numpy as np
import matplotlib.pyplot as plt

# training data
x = np.array([1.0, 2.0])
y = np.array([300.0, 500.0])

%matplotlib inline
```

## Cost Function

Cost measures how wrong our predictions are.

We use the Mean Squared Error cost:

J(w, b) = (1 / 2m) * sum((prediction - actual)^2)

Steps:
1. Compute prediction using w*x + b
2. Subtract actual value
3. Square the difference
4. Add all errors
5. Divide by 2m

If predictions are perfect, cost becomes 0.

```python
def compute_cost(x, y, w, b):
    """
    Compute Mean Squared Error Cost
    """
    m = len(x)

    total_error = 0

    for i in range(m):
        prediction = w * x[i] + b
        error = prediction - y[i]
        total_error += error**2

    cost = total_error / (2*m)

    return cost
```

## Testing the Cost Function

From manual calculation, we know:

w = 200  
b = 100  

should perfectly fit the data.

So cost should be 0.

```python
w = 200
b = 100

cost = compute_cost(x, y, w, b)
print("Cost:", cost)
```

## Visualizing the Linear Model

We now plot:

- Training data points
- The prediction line

This helps us visually understand how the model fits the data.

```python
def plot_line(x, y, w, b):
    plt.scatter(x, y, label="Training Data")

    x_line = np.linspace(0, 3, 100)
    y_line = w * x_line + b

    plt.plot(x_line, y_line, label="Model")
    plt.xlabel("Size")
    plt.ylabel("Price")
    plt.legend()
    plt.show()
```

```python
plot_line(x, y, w=200, b=100)
```

## Cost Intuition

What happens if w is incorrect?

If w is too small or too large:
- Predictions move away from actual values
- Errors increase
- Cost increases rapidly

Because errors are squared, cost grows fast.

```python
w_values = np.linspace(0, 400, 100)
cost_values = []

for w in w_values:
    cost_values.append(compute_cost(x, y, w, 100))

plt.plot(w_values, cost_values)
plt.xlabel("w")
plt.ylabel("Cost")
plt.title("Cost vs w")
plt.show()
```

## Conclusion

We learned:

- Linear Regression model: f(x) = w*x + b
- Cost measures prediction error
- Perfect fit leads to cost = 0
- Cost curve is convex
- Minimizing cost gives optimal parameters
