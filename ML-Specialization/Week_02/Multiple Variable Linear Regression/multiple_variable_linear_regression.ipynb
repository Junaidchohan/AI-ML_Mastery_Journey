## What is Multiple Variable Linear Regression

Multiple Variable Linear Regression is a machine learning algorithm used to predict a value using multiple input features.

Simple Linear Regression uses one feature:

y = wx + b

Multiple Linear Regression uses many features:

y = w1x1 + w2x2 + w3x3 + ... + b

Where:

x = input features  
w = weights  
b = bias  
y = predicted value

Example:
Predict house price using:
- size of house
- number of bedrooms
- age of house

```python
import numpy as np
```

We use NumPy because it allows efficient mathematical operations on vectors and matrices.
This helps implement machine learning algorithms faster.

## Creating a Simple Dataset

We will create a small dataset.

Features:
x1 = house size  
x2 = number of bedrooms

Target:
y = house price

```python
# Features: [size, bedrooms]
X = np.array([
    [100, 3],
    [120, 4],
    [80, 2],
    [150, 5]
])

# Target prices
y = np.array([200, 250, 160, 300])

print("Features:\n", X)
print("Target:\n", y)
```

## Understanding Features and Weights

Each feature has a weight.

Example:

x1 = house size  
x2 = bedrooms

Model:

y = w1*x1 + w2*x2 + b

Where:
w1 = importance of size
w2 = importance of bedrooms
b = bias

## Initialize Model Parameters

We start with some random weights and bias.
The model will later adjust them during training.

```python
# number of features
n = X.shape[1]
# print(X.shape)

# initialize weights
w = np.zeros(n)

# bias
b = 0

print("Initial weights:", w)
print("Initial bias:", b)
```

## Prediction Function

The prediction function calculates the output of the model.

Formula:

f(x) = w.x + b

where:
w.x = dot product of weights and features

```python
def predict(X, w, b):
    return np.dot(X, w) + b
```

## Making Predictions

We use the prediction function to estimate house prices using the current weights.

```python
predictions = predict(X, w, b)

print("Predictions:", predictions)
```

## Cost Function

The cost function measures how wrong our model predictions are.

We use Mean Squared Error (MSE):

J(w,b) = (1/2m) * Σ(prediction - actual)^2

Where:
m = number of training examples

```python
def compute_cost(X, y, w, b):
    m = len(y)
    predictions = np.dot(X, w) + b
    cost = (1/(2*m)) * np.sum((predictions - y)**2)
    return cost
```

## Initial Cost

We calculate how wrong the model is with the initial parameters.

```python
cost = compute_cost(X, y, w, b)
print("Initial Cost:", cost)
```

## Gradient Descent

Gradient Descent is an optimization algorithm used to minimize the cost function.

It updates weights and bias repeatedly to reduce prediction error.

Update rules:

w = w - α * derivative  
b = b - α * derivative

Where α is the learning rate.

```python
def gradient_descent(X, y, w, b, alpha, iterations):
    m = len(y)

    for i in range(iterations):

        predictions = np.dot(X, w) + b
        error = predictions - y

        dw = (1/m) * np.dot(X.T, error)
        db = (1/m) * np.sum(error)

        w = w - alpha * dw
        b = b - alpha * db

    return w, b
```

## Training the Model

Now we run gradient descent to update the weights and bias.

```python
alpha = 0.0001
iterations = 1000

w, b = gradient_descent(X, y, w, b, alpha, iterations)

print("Trained weights:", w)
print("Trained bias:", b)
```

## Final Predictions

After training, the model should produce better predictions.

```python
predictions = predict(X, w, b)

print("Final predictions:", predictions)
print("Actual values:", y)
```
