# Vectorization vs For Loop

Goal:

1. Implement dot product using a Python loop
2. Implement dot product using NumPy vectorization
3. Compare execution time
4. Understand why vectorization is faster

This is the foundation of Machine Learning computation.

```python
import numpy as np
import time
```

## Step 1: Create Large Vectors

We create two large vectors.
Large size helps us clearly see the speed difference.

```python
np.random.seed(1)

size = 5_000_000   # 5 million elements

a = np.random.rand(size)
b = np.random.rand(size)

print("Vector shape:", a.shape)
```

## Step 2: Dot Product Using Loop

We manually compute:

x = sum(a[i] * b[i])

This uses a Python loop.

```python
def my_dot(a, b):
    result = 0
    for i in range(len(a)):
        result += a[i] * b[i]
    return result

start = time.time()
loop_result = my_dot(a, b)
end = time.time()

print("Loop Result:", loop_result)
print("Loop Time (seconds):", end - start)
```

## Step 3: Dot Product Using Vectorization

Now we use NumPy's optimized function:

np.dot(a, b)

This avoids Python loops.

```python
start = time.time()
vector_result = np.dot(a, b)
end = time.time()

print("Vectorized Result:", vector_result)
print("Vectorized Time (seconds):", end - start)
```

## Step 4: Compare Results

Both results should be almost identical.

But execution time will be very different.

Vectorization is much faster because:

- It uses optimized C code internally
- It uses hardware parallelism
- It avoids slow Python loops

This is why Machine Learning relies heavily on vectorized operations.
