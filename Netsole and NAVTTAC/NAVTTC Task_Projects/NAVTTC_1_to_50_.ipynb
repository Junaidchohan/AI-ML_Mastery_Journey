## Task 28–40: NumPy

```python
# Task 28 - create numpy array
import numpy as np
arr = np.array([1, 2, 3])
print(arr)

# Task 29 - indexing
print(arr[0])

# Task 30 - slicing
print(arr[:2])

# Task 31 - data type
print(arr.dtype)

# Task 32 - copy array
copy_arr = arr.copy()
print(copy_arr)

# Task 33 - shape
print(arr.shape)

# Task 34 - reshape
arr2 = np.array([1, 2, 3, 4])
print(arr2.reshape(2, 2))

# Task 35 - iteration
for i in arr:
    print(i)

# Task 36 - joining arrays
print(np.concatenate((arr, arr)))

# Task 37 - splitting arrays
print(np.array_split(arr2, 2))

# Task 38 - searching
print(np.where(arr2 == 3))

# Task 39 - sorting
print(np.sort(arr2))

# Task 40 - random numbers
print(np.random.randint(1, 10, 3))
```

## Task 41–48: Pandas

## Task 49–50: Statistics

```python
# Task 49 - mean median mode
import statistics

nums = [1, 2, 3, 4, 4]

print(statistics.mean(nums))
print(statistics.median(nums))
print(statistics.mode(nums))

# Task 50 - standard deviation
print(statistics.stdev(nums))
```
