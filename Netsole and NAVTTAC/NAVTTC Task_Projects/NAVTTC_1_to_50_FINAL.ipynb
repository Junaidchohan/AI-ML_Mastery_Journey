## Task 28–40: NumPy

```
import numpy as np

arr=np.array([1,2,3])
print(arr)

print(arr[0])
print(arr[:2])
print(arr.dtype)

print(arr.copy())
print(arr.shape)

arr2=np.array([1,2,3,4])
print(arr2.reshape(2,2))

for i in arr: print(i)

print(np.concatenate((arr,arr)))
print(np.array_split(arr2,2))
print(np.where(arr2==3))
print(np.sort(arr2))
print(np.random.randint(1,10,3))
```

## Task 49–50: Statistics

```
import statistics

nums=[1,2,3,4,4]
print(statistics.mean(nums))
print(statistics.median(nums))
print(statistics.mode(nums))
print(statistics.stdev(nums))
```
