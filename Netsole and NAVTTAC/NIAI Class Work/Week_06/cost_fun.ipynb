```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 📊 Sample Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# 📌 Cost Function (MSE)
def compute_cost(m, b, X, y):
    n = len(X)
    predictions = m * X + b
    return (1/n) * np.sum((predictions - y) ** 2)

# -------------------------------
# 📉 Cost vs m (2D)
m_values = np.linspace(-1, 3, 100)
b_fixed = 1
cost_values = [compute_cost(m, b_fixed, X, y) for m in m_values]

# -------------------------------
# 🎥 3D Cost Surface
m_range = np.linspace(-1, 3, 50)
b_range = np.linspace(-2, 4, 50)

M, B = np.meshgrid(m_range, b_range)
Z = np.zeros_like(M)

for i in range(len(m_range)):
    for j in range(len(b_range)):
        Z[j, i] = compute_cost(M[j, i], B[j, i], X, y)

# -------------------------------
# 🎨 Plot All Together
fig = plt.figure(figsize=(18,5))

# 📊 Plot 1: Data + Regression Line
ax1 = fig.add_subplot(1, 3, 1)
ax1.scatter(X, y, label='Data Points')

m_example = 0.8
b_example = 1
y_pred = m_example * X + b_example

ax1.plot(X, y_pred, label=f'Line (m={m_example}, b={b_example})')
ax1.set_title("Linear Regression Fit")
ax1.set_xlabel("X")
ax1.set_ylabel("y")
ax1.legend()

# 📉 Plot 2: Cost vs m
ax2 = fig.add_subplot(1, 3, 2)
ax2.plot(m_values, cost_values)
ax2.set_title("Cost vs Slope (m)")
ax2.set_xlabel("m")
ax2.set_ylabel("Cost")

# 🎥 Plot 3: 3D Cost Surface
ax3 = fig.add_subplot(1, 3, 3, projection='3d')
ax3.plot_surface(M, B, Z)
ax3.set_title("3D Cost Function")
ax3.set_xlabel("Slope (m)")
ax3.set_ylabel("Intercept (b)")
ax3.set_zlabel("Cost")

plt.tight_layout()
plt.show()
```
