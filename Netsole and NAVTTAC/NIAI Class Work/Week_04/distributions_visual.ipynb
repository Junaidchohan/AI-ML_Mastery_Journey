## 1️⃣ Bernoulli Distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Example 1: Coin Toss
np.random.seed(0)
data = np.random.binomial(1, 0.5, 100)
bins=2
title="Bernoulli: Coin Toss"
xlabel="Outcome"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
# Example 2: Student Pass
data = np.random.binomial(1, 0.7, 100)
bins=2
title="Bernoulli: Student Pass"
xlabel="Outcome"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

## 7️⃣ t-Distribution

```python
import scipy.stats as stats

# Example 1
data = stats.t.rvs(df=5, size=1000)
bins=30
title="t-Distribution (df=5)"
xlabel="Value"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
# Example 2
data = stats.t.rvs(df=2, size=1000)
bins=30
title="t-Distribution (df=2)"
xlabel="Value"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```
