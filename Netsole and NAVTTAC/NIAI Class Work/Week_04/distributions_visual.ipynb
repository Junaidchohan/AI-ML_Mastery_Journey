# 📊 Distributions Visual



## 1️⃣ Bernoulli Distribution

# 📊 1. Bernoulli Distribution (Single Yes/No)
Idea:

Only one trial

Only two outcomes: 0 or 1

Example:

Toss a coin → Head (1) or Tail (0)

Student pass → Pass (1) or Fail (0)


One event → success or failure

```python
import numpy as np
import matplotlib.pyplot as plt

# Example 1: Coin Toss
np.random.seed(0)
data = np.random.binomial(1, 0.5, 100)
bins=2
title="Bernoulli: Coin Toss"
xlabel="Outcome"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
# Example 2: Student Pass
data = np.random.binomial(1, 0.7, 100)
bins=2
title="Bernoulli: Student Pass"
xlabel="Outcome"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

## 2️⃣ Binomial Distribution

# 📊 Binomial Distribution (Multiple Yes/No)
Idea:

Repeat Bernoulli many times

Count number of successes

Example:
Toss coin 10 times → how many heads?

20 students → how many pass?



Many trials → count successes

```python
# Example 1: 10 Coin Tosses
data = np.random.binomial(10, 0.5, 1000)
bins=10
title="Binomial: 10 Coin Tosses"
xlabel="Number of Heads"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
# Example 2: Student Passing
data = np.random.binomial(20, 0.6, 1000)
bins=10
title="Binomial: Student Passing"
xlabel="Number of Passes"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

## 3️⃣ Poisson Distribution

# 📊 Poisson Distribution (Rare Events)
Idea:

Counts how many times something happens

In a fixed time or space

Example:

Calls in 1 hour

Cars passing in 1 minute

Errors in a system



=> Counts rare events in fixed time

```python
# Example 1: Calls per Hour
data = np.random.poisson(3, 1000)
bins=10
title="Poisson: Calls per Hour"
xlabel="Number of Calls"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
# Example 2: Website Visits
data = np.random.poisson(5, 1000)
bins=10
title="Poisson: Website Visits"
xlabel="Visits"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

## 4️⃣ Exponential Distribution

# Exponential Distribution (Waiting Time)
Idea:

Measures time between events

Example:

Time until next call

Time until next customer arrives

```python
# Example 1: Time Between Calls
data = np.random.exponential(scale=2, size=1000)
bins=30
title="Exponential: Time Between Calls"
xlabel="Time"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
# Example 2: Customer Arrival
data = np.random.exponential(scale=1, size=1000)
bins=30
title="Exponential: Customer Arrival"
xlabel="Time"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

## 6️⃣ Normal Distribution

# Normal Distribution (Bell Curve)
Idea:

Most values are in the middle

Few are extreme

Example:

Human height

IQ scores

Shape:

Bell-shaped curve



=> Real-life natural data pattern

```python
# Example 1: Student Marks
data = np.random.normal(70, 10, 1000)
bins=30
title="Normal: Student Marks"
xlabel="Marks"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

## 7️⃣ t-Distribution

# t-Distribution (Small Data)
Idea:

Same as normal BUT

Used when sample size is small (< 30)

Example:

You only test 10 students

Small experiment data

Special:

Wider than normal (more uncertainty)


Used when data is small

```python
import scipy.stats as stats

# Example 1
data = stats.t.rvs(df=5, size=1000)
bins=30
title="t-Distribution (df=5)"
xlabel="Value"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
# Example 2
data = stats.t.rvs(df=2, size=1000)
bins=30
title="t-Distribution (df=2)"
xlabel="Value"

plt.figure(figsize=(6,4))
plt.hist(data, bins=bins)
plt.title(title)
plt.xlabel(xlabel)
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```
