# 📊 Distributions Explained (Simple + Visual)

## 1️⃣ Bernoulli Distribution
👉 Only two outcomes (0 or 1)


```python
import numpy as np
import matplotlib.pyplot as plt

# Example 1: Coin Toss
np.random.seed(0)
data = np.random.binomial(1, 0.5, 100)

plt.hist(data, bins=2)
plt.title("Bernoulli: Coin Toss")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```

```python
# Example 2: Student Pass
data = np.random.binomial(1, 0.7, 100)

plt.hist(data, bins=2)
plt.title("Bernoulli: Student Pass")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```

## 2️⃣ Binomial Distribution
👉 Count successes in multiple trials


```python
# Example 1: 10 Coin Tosses
data = np.random.binomial(10, 0.5, 1000)

plt.hist(data, bins=10)
plt.title("Binomial: 10 Coin Tosses")
plt.xlabel("Number of Heads")
plt.ylabel("Frequency")
plt.show()
```

```python
# Example 2: Student Passing
data = np.random.binomial(20, 0.6, 1000)

plt.hist(data, bins=10)
plt.title("Binomial: Student Passing")
plt.xlabel("Number of Passes")
plt.ylabel("Frequency")
plt.show()
```

## 3️⃣ Poisson Distribution
👉 Count rare events in fixed time


```python
# Example 1: Calls per Hour
data = np.random.poisson(3, 1000)

plt.hist(data, bins=10)
plt.title("Poisson: Calls per Hour")
plt.xlabel("Number of Calls")
plt.ylabel("Frequency")
plt.show()
```

```python
# Example 2: Website Visits
data = np.random.poisson(5, 1000)

plt.hist(data, bins=10)
plt.title("Poisson: Website Visits")
plt.xlabel("Visits")
plt.ylabel("Frequency")
plt.show()
```

## 4️⃣ Exponential Distribution
👉 Time between events


```python
# Example 1: Time Between Calls
data = np.random.exponential(scale=2, size=1000)

plt.hist(data, bins=30)
plt.title("Exponential: Time Between Calls")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()
```

```python
# Example 2: Customer Arrival Time
data = np.random.exponential(scale=1, size=1000)

plt.hist(data, bins=30)
plt.title("Exponential: Customer Arrival")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()
```

## 5️⃣ Uniform Distribution
👉 Equal probability everywhere


```python
# Example 1: Random Numbers
data = np.random.uniform(0, 10, 1000)

plt.hist(data, bins=20)
plt.title("Uniform: 0 to 10")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Example 2: Random Scores
data = np.random.uniform(50, 100, 1000)

plt.hist(data, bins=20)
plt.title("Uniform: Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

## 6️⃣ Normal Distribution
👉 Natural real-life data


```python
# Example 1: Student Marks
data = np.random.normal(70, 10, 1000)

plt.hist(data, bins=30)
plt.title("Normal: Student Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
# Example 2: Human Height
data = np.random.normal(170, 10, 1000)

plt.hist(data, bins=30)
plt.title("Normal: Height")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.show()
```

## 7️⃣ t-Distribution
👉 Used for small samples


```python
import scipy.stats as stats

# Example 1
data = stats.t.rvs(df=5, size=1000)

plt.hist(data, bins=30)
plt.title("t-Distribution (df=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Example 2
data = stats.t.rvs(df=2, size=1000)

plt.hist(data, bins=30)
plt.title("t-Distribution (df=2)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
