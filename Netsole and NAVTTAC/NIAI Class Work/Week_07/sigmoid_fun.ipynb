```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Log loss functions
def loss_y1(p):
    return -np.log(p)          # when y = 1

def loss_y0(p):
    return -np.log(1 - p)      # when y = 0

# Generate z values
z = np.linspace(-10, 10, 200)
p = sigmoid(z)

# Compute losses
loss1 = loss_y1(p)
loss0 = loss_y0(p)

# Plot
plt.figure()

# Sigmoid curve
plt.plot(z, p, linestyle='--', label="Sigmoid (Probability)")

# Log loss curves
plt.plot(z, loss1, label="Loss when y = 1  →  -log(p)")
plt.plot(z, loss0, label="Loss when y = 0  →  -log(1 - p)")

# Labels
plt.xlabel("z")
plt.ylabel("Value")
plt.title("Sigmoid Function and Log Loss")

# Limits (important for clarity)
plt.ylim(0, 10)

# Legend & grid
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate z values
z = np.linspace(-10, 10, 200)
p = sigmoid(z)

# Cost for y = 0
cost = -np.log(1 - p)

# Plot
plt.figure()

# Plot cost function
plt.plot(z, cost, label="Cost when y = 0  →  -log(1 - h(x))")

# Optional: also plot sigmoid for reference
plt.plot(z, p, linestyle='--', label="Sigmoid (h(x))")

# Labels
plt.xlabel("z")
plt.ylabel("Cost")
plt.title("Logistic Regression Cost Function (y = 0)")

# Limit for better view
plt.ylim(0, 10)

# Legend & grid
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate z values
z = np.linspace(-10, 10, 200)
p = sigmoid(z)

# Assume true label y = 1
y = 1

# MSE Loss
mse = (y - p) ** 2

# Cross-Entropy Loss
ce = - (y * np.log(p) + (1 - y) * np.log(1 - p))

# Plot
plt.figure()

# Plot both losses
plt.plot(z, mse, label="MSE Loss (y=1)")
plt.plot(z, ce, label="Cross-Entropy Loss (y=1)")

# Labels
plt.xlabel("z")
plt.ylabel("Loss")
plt.title("MSE vs Cross-Entropy (y = 1)")

# Limit for better visualization
plt.ylim(0, 5)

# Legend & grid
plt.legend()
plt.grid()

plt.show()
```
